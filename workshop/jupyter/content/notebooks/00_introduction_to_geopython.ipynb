# GeoPython Workshop (english)


![GeoPython logo](images/geopython-logo.png)

Welcome to the GeoPython workshop. 
This workshop should give you a brief overview about how to use Python when working 
with geospatial data.

## Working environment

This workshop will use Jupyter Notebooks. 
As the installation of all dependencies on all platforms (Win, Mac, Linux)
can be quite involved we have contained Jupyter and all dependencies within a Docker Image.

Also the various geospatial web services like pygeoapi and pycsw, are provided by Docker Images.

### Requirements

The main requirement is to have Docker and Docker Compose installed.

### Optional Requirements

You may optionally install [QGIS](http://qgis.org) as a GIS data viewer. 
QGIS is a multiplatform open source desktop application for geospatial data.

## Workshop data

### Input data

Data are located in each directory, so you should be able to access 
them by using `./data/file.xyz` 

### Ouput data

Outputs should be located in `/data/` directory.


## Motivation

Why to use Python anyway?

* Python makes fun
* You don't really have to learn Python - you simply write it
* It is the most geo-positive programming language around
* Batteries included - many tools at hand ([GDAL](http://gdal.org), [PROJ4](http://proj4.org), [Shapely](http://toblerity.org/shapely/manual.html), [Fiona](http://toblerity.org/fiona/manual.html), [Rasterio](https://rasterio.readthedocs.io/en/latest/intro.html), MapServer Python [MapScript](http://mapserver.org/mapscript/python.html), GeoServer [gsconfig](https://github.com/boundlessgeo/gsconfig), [OWSLib](https://geopython.github.io/OWSLib/), [PyWPS](http://pywps.org), [pycsw](http://pycsw.org), [...](https://geopython.qgis.io))
* Support in desktop programs ([GRASS GIS](http://grass.osgeo.org) - PyGRASS, [Esri ArcGIS](http://esri.com) - arcpy, [QGIS](http://qgis.org) - PyQGIS, ...)
* GIS is fun, everybody likes maps in the computer
* ...
